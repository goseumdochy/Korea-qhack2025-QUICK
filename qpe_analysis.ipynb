```python
import numpy as np
import matplotlib.pyplot as plt

def plot_fn(n, num_points=1000):
    t = np.linspace(0.0001, 1-0.0001, num_points)
    epsilon = 1e-10  

    numerator = 1 - np.cos(2 * np.pi * 2**n * t)
    denominator = 1 - np.cos(2 * np.pi * t) + epsilon 

    f_t = (1 / 4**n) * (numerator / denominator)

    plt.figure(figsize=(8, 4))
    plt.plot(t, f_t, label=f"$n = {n}$")
    plt.title(r"$f_n(t) = \frac{1}{4^n} \cdot \frac{1 - \cos(2\pi \cdot 2^n t)}{1 - \cos(2\pi t)}$")
    plt.xlabel("t")
    plt.ylabel(f"$f_{n}(t)$")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_fn(n=3)
```
